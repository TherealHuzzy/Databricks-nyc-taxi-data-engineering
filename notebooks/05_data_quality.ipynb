# 🛡️ Data Quality Monitoring

## Objective

This notebook performs automated data quality checks on the NYC Taxi dataset after data has been transformed into the Gold layer.

The purpose is to ensure that analytical datasets are accurate, complete, and reliabe before they are consumed by dashboards and downstream applications.

## Data Quality Checks

The following validations are performed:

- ✅ Check for duplicate records
- ✅ Check for missing values
- ✅ Verify trip distance is greater than zero
- ✅ Verify fare amount is non-negative
- ✅ Validate passenger count is greater than zero
- ✅ Generate a summary report of all validation results

```python
daily_df = spark.read.table("workspace.default.daily_summary")
hourly_df = spark.read.table("workspace.default.hourly_summary")
vendor_df = spark.read.table("workspace.default.vendor_summary")
pickup_df = spark.read.table("workspace.default.pickup_summary")
payment_df = spark.read.table("workspace.default.payment_summary")
```

```python
# Ensuring tables are not empty
assert daily_df.count() > 0, "Table daily_summary is empty"
assert hourly_df.count() > 0, "Table hourly_summary is empty"
assert vendor_df.count() > 0, "Table vendor_summary is empty"
assert pickup_df.count() > 0, "Table pickup_summary is empty"
assert payment_df.count() > 0, "Table payment_summary is empty"
```

```python
# Check for null values in trip_date column
from pyspark.sql.functions import col

assert daily_df.filter(col("trip_date").isNull()).count() == 0, "Null trip date found"
```

```python
# trips must be positive
assert daily_df.filter(col("total_trips") <= 0).count() == 0, \
    "Invalid trip counts detected."
```

```python
# Revenue must be positive
assert daily_df.filter(col("daily_revenue") <= 0).count() == 0, \
    "Invalid revenue detected."
```

```python
# Trip distance must be positive
assert daily_df.filter(col("total_distance") <= 0).count() == 0, \
    "Invalid trip distance detected."
```

```python
# Payment types should be valid
assert payment_df.filter(~col("payment_type").isin([1,2,3,4])).count() == 0, \
    "Unxpected payment type detected."
```

```python
# Print a success message
print("✅ All data quality checks passed susccessfully")
```
