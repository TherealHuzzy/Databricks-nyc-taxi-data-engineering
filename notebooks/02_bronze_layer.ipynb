# Bronze Layer

## Objective
Read the January 2026 NYC Yellow Taxi dataset and persist it as a Delta table in the bronze layer without applying transformations.

## Input
- yellow_tripdata_2026-01.parquet

## Output
- Bronze Delta table

## Expected Result
The raw dataset is stored in Delta format and is ready for downstream processing.

Reading the raw parquet file

```python
df = spark.read.parquet(
    "/Volumes/workspace/default/nyc_raw_data/yellow_tripdata_2026-01.parquet"
)
```

Writing it as a Delta table

```python
df.write.format("delta").mode("overwrite").save(
    "/Volumes/workspace/default/nyc_raw_data/bronze")
```
